# CatBoost

```python
from catboost import CatBoostClassifier
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

## Importando os dados

```python
Y_treino = pd.read_csv('/home/lucasqueiros/DataScience/Projeto-Classificacao/data/processed/Y_treino.csv', sep=',', encoding='utf-8')
Y_teste = pd.read_csv('/home/lucasqueiros/DataScience/Projeto-Classificacao/data/processed/Y_teste.csv', sep=',', encoding='utf-8')

X_treino = pd.read_csv('/home/lucasqueiros/DataScience/Projeto-Classificacao/data/processed/X_treino.csv', sep=',', encoding='utf-8')
X_teste = pd.read_csv('/home/lucasqueiros/DataScience/Projeto-Classificacao/data/processed/X_teste.csv', sep=',', encoding='utf-8')
```

```python
Y_teste = Y_teste.iloc[:, 1].values
Y_treino = Y_treino.iloc[:, 1].values

X_esc_teste = X_teste.iloc[:, 1:].values
X_esc_treino = X_treino.iloc[:, 1:].values
```

### Treinando modelo com dados escalonados

```python
catboost = CatBoostClassifier(task_type='CPU', iterations=100, learning_rate=0.1, depth = 8, random_state = 5,
                              eval_metric="Accuracy")
```

```python
catboost.fit(X_treino, Y_treino, plot=True, eval_set=(X_teste, Y_teste))
```

```python
previsoes_cat = catboost.predict(X_teste)
previsoes_cat
```

```python
Y_teste
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(Y_teste, previsoes_cat) * 100.0))
```

## Análise com dados de treino

```python
previsoes_treino = catboost.predict(X_treino)
previsoes_treino
```

```python
Y_treino.shape, previsoes_treino.shape
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(Y_treino, previsoes_treino) * 100.0))
```

```python
confusion_matrix(Y_treino, previsoes_treino)
```

## Validação Cruzada

```python
# Separando os dados em folds
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
modelo = CatBoostClassifier(task_type='CPU', iterations=100, learning_rate=0.1, depth = 8, random_state = 5,
                              eval_metric="Accuracy")
```

```python
resultado = cross_val_score(modelo, X_treino, Y_treino, cv = kfold)
resultado
```
